```python
# Step 1: Correlation Checks
# Correlation between staffing and wait

import pandas as pd

df = pd.read_csv("C:/Users/pooja/Documents/Job Applications/Data Analyst/Projects/Healthcare/ER-Wait-Time-Project/data/cleaned/df_cleaned.csv")

# Correlation: Nurse-to-Patient Ratio vs Total Wait
corr_nurse = df[['nurse_to_patient_ratio','total_wait_time_min']].corr().iloc[0,1]

# Correlation: Specialist Availability vs Total Wait
corr_specialist = df[['specialist_availability','total_wait_time_min']].corr().iloc[0,1]

print("Correlation (Nurse Ratio vs Wait):", corr_nurse)
print("Correlation (Specialist Availability vs Wait):", corr_specialist)
```

```python
# Step 2: Identify Bottlenecks
# Percentage of wait time spent in each stage

avg_stage_pct = df[['reg_pct','triage_pct','medical_pct']].mean()
print("Average % wait by stage:\n", avg_stage_pct)
```

```python
# Step 3: Urgency & Outcome Impact

urgency_wait_check = df.groupby('urgency_level')['total_wait_time_min'].mean()
print("Wait time by urgency:\n", urgency_wait_check)

satisfaction_corr = df[['total_wait_time_min','patient_satisfaction']].corr().iloc[0,1]
print("Correlation Total Wait vs Satisfaction:", satisfaction_corr)
```
