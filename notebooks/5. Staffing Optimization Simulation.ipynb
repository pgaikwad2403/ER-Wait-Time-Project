```python
# Step 1: Identify Peak Hours (Evening)

import pandas as pd
df = pd.read_csv("C:/Users/pooja/Documents/Job Applications/Data Analyst/Projects/Healthcare/ER-Wait-Time-Project/data/cleaned/df_cleaned.csv")

peak_hours_df = df[df['time_of_day']=='Evening'].copy()
```

```python
# Step 2: Simulate Staff Increase (Reduce Wait by 15%)

peak_hours_df['simulated_wait_min'] = peak_hours_df['total_wait_time_min'] * 0.85
```

```python
# Step 3: Compare Metrics

avg_before = peak_hours_df['total_wait_time_min'].mean()
avg_after = peak_hours_df['simulated_wait_min'].mean()

pct_patients_over_60_before = (peak_hours_df['total_wait_time_min'] > 60).mean()*100
pct_patients_over_60_after = (peak_hours_df['simulated_wait_min'] > 60).mean()*100

print(f"Average Wait Before: {avg_before:.2f} min")
print(f"Average Wait After: {avg_after:.2f} min")
print(f"% Patients >60 min Before: {pct_patients_over_60_before:.2f}%")
print(f"% Patients >60 min After: {pct_patients_over_60_after:.2f}%")
```
