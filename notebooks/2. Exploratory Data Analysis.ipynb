```python
# Step 1: Overall Metrics

import pandas as pd
df = pd.read_csv("C:/Users/pooja/Documents/Job Applications/Data Analyst/Projects/Healthcare/ER-Wait-Time-Project/data/cleaned/df_cleaned.csv")
print("Average wait:", df['total_wait_time_min'].mean())
print("Median wait:", df['total_wait_time_min'].median())
print("Patients leaving without seen:", (df['patient_outcome']=='Left Without Being Seen').mean()*100)
```

```python
# Step 2: Peak Hours & Days

# Time of Day
df.groupby('time_of_day')['total_wait_time_min'].mean().plot(kind='bar')

# Day of Week
df.groupby('day_of_week')['total_wait_time_min'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
# Step 3: Urgency Level Impact

df.groupby('urgency_level')['total_wait_time_min'].mean().plot(kind='bar')
```

```python
# Step 4: Staffing & Hospital Effect

df.groupby('nurse_to_patient_ratio')['total_wait_time_min'].mean().plot()
df.groupby('specialist_availability')['total_wait_time_min'].mean().plot()
```
